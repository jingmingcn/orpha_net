```python
import json
import pandas as pd

orphacode_list = []
name_list = []
alternateName_list = []
epidemiology_list = []
description_list = []
OMIM_list = []
MeSH_list = []
UMLS_list = []
SNOMED_CT_list = []
ICD_10_list = []
Medline_list = []
MedDRA_list = []
ICD_11_list = []
GARD_list = []
MONDO_list = []
with open('orpha.json', 'r') as file:
    json_object = json.load(file)
    for item in json_object:
        data = item['data']
        
        code = {'OMIM' :'', 
            'MeSH' :'',
            'UMLS' :'',
            'SNOMED CT' :'',
            'ICD-10'  :'',
            'Medline' :'',
            'MedDRA' :'',
            'ICD-11' :'',
            'GARD' :'',
            'MONDO' :'',
        }
        for e in data['code']:
            code[e['codingSystem']] = e['codeValue']

        orphacode_list.append(item['orphacode'])
        name_list.append(data['name']['textValue'])
        alternateName_list.append(data['alternateName'])
        epidemiology_list.append(data['epidemiology'])
        OMIM_list.append(code['OMIM'])
        MeSH_list.append(code['MeSH'])
        UMLS_list.append(code['UMLS'])
        SNOMED_CT_list.append(code['SNOMED CT'])
        ICD_10_list.append(code['ICD-10'])
        Medline_list.append(code['Medline'])
        MedDRA_list.append(code['MedDRA'])
        ICD_11_list.append(code['ICD-11'])
        GARD_list.append(code['GARD'])
        MONDO_list.append(code['MONDO'])
        description_list.append(data['description'])
    
    df = pd.DataFrame({
        'OrphaCode': orphacode_list,
        'Name': name_list,
        'Alternate Name': alternateName_list,
        'Epidemiology': epidemiology_list,
        'OMIM': OMIM_list,
        'MeSH': MeSH_list,
        'UMLS': UMLS_list,
        'SNOMED CT': SNOMED_CT_list,
        'ICD-10': ICD_10_list,
        'Medline': Medline_list,
        'MedDRA': MedDRA_list,
        'ICD-11': ICD_11_list,
        'GARD': GARD_list,
        'MONDO': MONDO_list, 
        'Description': description_list
    })

    df.to_csv('orpha.csv', index=False)
```
